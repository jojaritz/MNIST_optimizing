```python
import kagglehub
import os
import pandas as pd
import torch



# Download latest version

path = kagglehub.dataset_download("shashanknecrothapa/ames-housing-dataset")



print("Path to dataset files:", path)

print("Files in this directory:", os.listdir(path))
```

```python
csv_path = os.path.join(path, "AmesHousing.csv")

df=pd.read_csv(csv_path)

print(df.columns)
print(df.tail(20))
```

```python
features = ['Gr Liv Area', 'Lot Area', 'Year Built']
target = ['SalePrice']

X_raw = df[features].copy()
y_raw = df[target].copy()



X_norm = (X_raw - X_raw.mean()) / X_raw.std()
y_scaled = y_raw / 100000


X = torch.tensor(X_norm.values, dtype=torch.float32)
y = torch.tensor(y_scaled.values, dtype=torch.float32)

print(X.shape)
print(y.shape)
```

```python
W = torch.randn(3, 1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)


def forward_pass(X_data):
  y_pred = torch.matmul(X_data, W) + b
  return y_pred
predictions = forward_pass(X)
print(predictions.shape)
print(predictions.max())
```
